# [The Nature Conservancy Fisheries Monitoring ](https://www.kaggle.com/c/the-nature-conservancy-fisheries-monitoring/data)

In this competition, The Nature Conservancy asks you to help them detect which species of fish appears on a fishing boat, based on images captured from boat cameras of various angles.  

### Your goal is to predict the likelihood of fish species in each picture.

**Eight** target categories are available in this dataset: **Albacore tuna, Bigeye tuna, Yellowfin tuna, Mahi Mahi, Opah, Sharks, Other** (meaning that there are fish present but not in the above categories), and **No Fish** (meaning that no fish is in the picture). 
Each image has only one fish category, except that there are sometimes very small fish in the pictures that are used as bait.

```python
from fastai import *
from fastai.vision import *
```

```python
PATH = Path('dataset')
PATH.ls()
```

```python
bs = 32
```

```python
def class_checker(path, train_folder="train",valid_folder="valid"):
    notInTrain = []
    notInValid = []
    path = Path(path)
    train_check = (path/train_folder).ls()
    valid_check = (path/valid_folder).ls()
    for i in train_check:
        if i not in valid_check: notInValid.append(i)
    for i in valid_check:
        if i not in train_check: notInTrain.append(i)
    return notInTrain,notInValid

class_checker(PATH)
```

```python
data = ImageDataBunch.from_folder(PATH, train="train", valid='valid', test="test", ds_tfms = get_transforms(), size=224, bs=bs)
```

```python
data.normalize(imagenet_stats)
```

```python
doc(data.show_batch)
```

```python
data.show_batch(rows = 3, figsize=(12,10))
```

## Training: resnet50

```python
learn = create_cnn(data, models.resnet50, metrics=error_rate)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.load('stage-1')
```

#### Running one cycle policy with lr = 6e-3

```python
learn.fit_one_cycle(10, max_lr=slice(6e-3))
```

```python
learn.save('resnet50_stage_1')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9, figsize=(12,12))
```

```python
doc(interp.plot_top_losses)
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
learn.unfreeze()
learn.fit_one_cycle(2)
```

```python
learn.load('resnet50_stage_1')
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, max_lr=slice(6e-6,1e-4))
```

```python
learn.save('resnet50_stage_2')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(8, max_lr=slice(2e-6,1e-4))
```

```python
learn.save('resnet50_stage_3_full')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
interp.most_confused(min_val=2)
```

#### model is overfitted, rolling back to 1st model, unfreeze and train

```python
learn.load('resnet50_stage_1')
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(5, max_lr=slice(1e-5,1e-4))
```

```python
learn.save('resnet50_stage_2_full')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```
